# Exercise 1.1

## Solution 1 - Multiple If...continue
This code effectively says:
> "For each number, if multiple of 4 then skip, if end in 7 then skip, else print the number"

```python
for n in range(1,1001):

  if n % 4 == 0:
    continue

  if str(n)[-1] == '7':
    continue

  print(n)
```

## Solution 2 - Single If
This code effectively says:
> "For each number, if not a multiple of 4 or not end in 7, print the number"

```python
for n in range(1,1001):
  if n % 3 != 0 and str(n)[-1] != '4':
    print(n)
```

# Exercise 1.2

```python
units = input('Enter M to use metric or I to use imperial units ... ')

# Create a bool to remember if the user input metric
# We are converting the units to upper case, in case they input m rather than M
is_metric = units.upper() == 'M'

# Get the weight units as string so we can use them in the input question
if is_metric:
  weight_unit = 'kg'
  height_unit = 'm'
else:
  weight_unit = 'lbs'
  height_unit = 'in'

# Ask for the inputs - we have added the selected units to the input question
weight = input(f'Input weight ({weight_unit})... ')
height = input(f'Input height ({height_unit})... ')

# Cast the string inputs as floats
weight = float(weight)
height = float(height)

# Compute BMI, adjusting the calculation if imperial units are used
if is_metric:
  bmi = weight / height ** 2
else:
  bmi = 703 * weight / height ** 2

# Print the result value
print('bmi =', bmi)

# Print the category
if bmi < 18.5:
  print('Underweight')
elif bmi < 25:
  print('Healthy')
elif bmi < 30:
  print('Overweight')
elif bmi < 40:
  print('Obese')
else:
  print('Severely Obese')

# Generate weight range - note we need to cast the float values back to integers
weights = range(int(0.5 * weight), int(1.5 * weight), 10)

# Print the table header
print("height, weight, bmi, category")

# For each weight we want to add to the table
for w in weights:

  # Calculate the BMI
  if is_metric:
    bmi = w / height ** 2
  else:
    bmi = 703 * w / height ** 2

  # Calculate the category
  if bmi < 18.5:
    bmi_cat = 'Underweight'
  elif bmi < 25:
    bmi_cat = 'Healthy'
  elif bmi < 30:
    bmi_cat = 'Overweight'
  elif bmi < 40:
    bmi_cat = 'Obese'
  else:
    bmi_cat = 'Severely Obese'

  # Print the row of the table using the f-string syntax
  print(f"{height} {height_unit}, {w} {weight_unit}, {bmi}, {bmi_cat}")
```
